```python
import numpy as np
import pandas as pd
# making a Dataframe from dictionary
students = {
    "name": ['Maya','Ankita','Preet','Maryam','Sahil','Jaishwal'],
    "age": [23,55,64,32,20,16],
    "MaxEducation": ['Bachelor','PHD','PHD','Masters','Undergrad','Intermediate'],
    "course": ['BSCS','Maths','English','Economics','BBA','ICS'],
    "Institute": ['Muhammad Ali Jinnah University','IBA','IBA','King Saud University','Habib University','Bahria College'],
    "email": ['Maya.edu@gmail.com','Ankita.edu@gmail.com','Preet.edu@gmail.com','Maryam.edu@gmail.com','Sahil.edu@gmail.com','Jaishwal.edu@gmail.com']}
students

studentInfo = pd.DataFrame(students)
studentInfo    
```

```python
studentInfo.index = [102134232011,102134232012,102134232013,102134232014,102134232015,102134232016]
studentInfo
```

```python
# agar ye method use krengy tw sabko change krna hoga
# column name changing method 1
studentInfo.columns = [
    "Name",
    "Age",
    "MaxEducation",
    "Course",
    "Institute",
    "Email"
]
studentInfo
```

```python
# column name renaming method 2
studentInfo.rename(columns={
    "MaxEducation": "Higher Education",
    "Course": "Degree"}, inplace = True)
```

```python
import numpy as np
age = np.array([12,33,44,55])
name = np.array(["maya","tina","aleena","preet"])
# np.nditer() does not accept two separate arguments like this.
for i,k in np.nditer(name,age):
    print(f'My name is {i} and my age is {k}.')
```

```python
import numpy as np
age = np.array([12,33,44,55])
name = np.array(["maya","tina","aleena","preet"])
# np.nditer() does not accept two separate arguments like this.
for i,k in zip(name,age):
    print(f'My name is {i} and my age is {k}.')
```
