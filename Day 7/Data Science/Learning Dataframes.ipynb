```python
import numpy as np
age = np.array([12,33,44,55])
name = np.array(["maya","tina","aleena","preet"])
# np.nditer() does not accept two separate arguments like this.
for i,k in np.nditer(name,age):
    print(f'My name is {i} and my age is {k}.')
```

```python
import numpy as np
age = np.array([12,33,44,55])
name = np.array(["maya","tina","aleena","preet"])
# np.nditer() does not accept two separate arguments like this.
for i,k in zip(name,age):
    print(f'My name is {i} and my age is {k}.')
```
